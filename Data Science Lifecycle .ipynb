# Missing values Analysis

```python
features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() >1]
```

```python
features_nan
```

```python
for feature in features_nan:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), 'Missing value')
```

```python
for feature in features_nan:
    data =dataset.copy()
    
    data[feature] = np.where(dataset[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].mean().plot.bar()
    plt.title(feature)
    plt.show()
    
```
